```python
#my code. unnessary use list.same thing can be done with the  help of string  
def compress(st):
    count = 1
    li = []
    for i in range(1,len(st)):
        if st[i] ==  st[i-1]:
            count+=1
        else:
            num = str(count)
            con = st[i-1] + num
            li.append(con)
            count = 1
        i+=1
    num = str(count)
    con = st[i-1] + num
    li.append(con)
    return ("".join(li))
```

```python
st = 'AAASSSWqQEEEESSS'
x = compress(st)
print(x)
```
