```python
def permute(s):
    out = []
    
    # Base Case
    if len(s) == 1:
        print("value of s is:-",s)
        out = [s]
        
    else:
        # For every letter in string
        for i, let in enumerate(s):
            
            # For every permutation resulting from Step 2 and 3 described above
            #print(s[:i] + s[i+1:])
            for perm in permute(s[:i] + s[i+1:]):
                
                #print("letter is",let)
                #print("perm is",perm)
                
                # Add it to output
                out += [let + perm]
                print(out)

    return out
```
